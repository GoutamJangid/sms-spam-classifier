### 3. Text Preprocessing


* **Lower Case**
  * *Converting all text to lowercase to treat words like "Free" and "free" as the same.*

* **Tokenization**
  * *Breaking down sentences into individual words (tokens) for the model to process.*

* **Removing special characters**
  * *Eliminating punctuation and non-alphanumeric symbols to reduce noise in the data.*

* **Removing stop words and punctuation**
  * *Filtering out common words (e.g., "the", "is", "a") that don't add significant meaning.*

* **Stemming**
  * *Reducing words to their root form (e.g., "loving" -> "love") to group related words together.*

```python
import nltk
nltk.download('stopwords')
```

```python
#stopwords
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
print(stopwords)
```

```python
# punctuations
import string
punctuations = string.punctuation
print(punctuations)
```

```python
# stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
ps.stem('stemming') 
```

```python
def transform_text(text): 

    #converting into lower case 
    text = text.lower()

    # breaking into tokens
    text = nltk.wordpunct_tokenize(text)

    # remove special chars
    y = [] 
    for i in text:
        if i.isalnum(): # only appending alphanumeric characters (letters and numbers)
            y.append(i)

    text = y[:]
    y.clear()

    #removing stopwords and punctuation
    for i in text:
        if i not in stopwords and punctuations:
            y.append(i)

    text = y[:]
    y.clear()

    # stemming 
    for i in text:
        y.append(ps.stem(i))
    
    return " ".join(y) #Join the list of words back into a single string since, vectorizers expect string input, not a list.
```

```python
transform_text('Are you enjoying this project or not @$@#(&(* ?')
```

```python
# apply to our dataframe
df['transformed_text'] = df['text'].apply(transform_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score

# Vectorization and Data Splitting 
tfidf = TfidfVectorizer(max_features=3000)
X = tfidf.fit_transform(df['transformed_text']).toarray()
y = df['target'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Initialize All Models 
mnb = MultinomialNB()
bnb = BernoulliNB()
lrc = LogisticRegression(solver='liblinear', penalty='l1')
svc = SVC(kernel='sigmoid', gamma=1.0)
dtc = DecisionTreeClassifier(max_depth=5)
rfc = RandomForestClassifier(n_estimators=50, random_state=2)

# Train, Predict, and Evaluate 
models = {
    "Multinomial Naive Bayes": mnb,
    "Bernoulli Naive Bayes": bnb,
    "Logistic Regression": lrc,
    "Support Vector Machine": svc,
    "Decision Tree": dtc,
    "Random Forest": rfc
}

results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    results.append({'Model': name, 'Accuracy': acc, 'Precision': prec})

# Display Results 
results_df = pd.DataFrame(results)
print(results_df)
```

```python

performance_df = pd.melt(results_df, id_vars=['Model'], value_vars=['Accuracy', 'Precision'])

performance_df.rename(columns={
    'Model': 'Algorithm',
    'variable': 'Metric',
    'value': 'Score'
}, inplace=True)

sns.catplot(data=performance_df, x='Algorithm', y='Score', hue='Metric',
            kind='bar', height=4, aspect=0.8)

plt.ylim(0.8, 1.0)
plt.xticks(rotation='vertical')
plt.title('Model Performance Comparison')
plt.show()
```

### 5.Model Improvement
* **Experiment A: Different Vectorizer (CountVectorizer)**

* **Experiment B: Ensemble Method (VotingClassifier)**

* **Experiment C: Use N-grams**

```python
# Experiment A: Different Vectorizer (CountVectorizer)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score

# Vectorize the text data
cv = CountVectorizer(max_features=3000)
X = cv.fit_transform(df['transformed_text']).toarray()
y = df['target'].values

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Train the model
model = MultinomialNB()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("CountVectorizer + MultinomialNB")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
```

```python
# clearly its not better 
```

```python
# Experiment B: Ensemble Method (VotingClassifier)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score

# TF-IDF vectorized data assumed to be already split: X_train, X_test, y_train, y_test

# Set up individual models
nb = MultinomialNB()
lr = LogisticRegression(solver='liblinear', penalty='l1')
svm = SVC(kernel='sigmoid', gamma=1.0, probability=True)

# Create the voting classifier
voting_clf = VotingClassifier(
    estimators=[('NB', nb), ('LR', lr), ('SVM', svm)],
    voting='soft'
)

# Train and evaluate
voting_clf.fit(X_train, y_train)
y_pred = voting_clf.predict(X_test)

print("Voting Classifier Results")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
```

```python
# Experiment C: Use N-grams
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score

# TF-IDF vectorizer with unigrams and bigrams
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X = tfidf.fit_transform(df['transformed_text']).toarray()
y = df['target'].values

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Train the model
model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Results
print("MultinomialNB with TF-IDF (1-gram + 2-gram)")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
```

### Model Improvement Results
